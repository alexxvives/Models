# Logistic Regression
Logistic regression is a classification algorithm. It is used for tasks involving two or more classes based on a set of independent variables.

There are three types of Logistic Regression:
* Binary logistic regression (covered in notebook)
* Multinomial logistic regression
* Ordinal logistic regression

*This notebook will only cover binary logistic regression, because multinomial and ordinal logistic regressions return decision boundaries dividing each class from all the others (multiple replication of the binary logistic regression procedure).*

### Binary Logistic Regression
Binary logistic regression is a statistical model used for binary classification tasks, where the goal is to predict the probability of an example belonging to one of two classes. It is a type of generalized linear model that uses a logistic function to model the probability of a binary outcome.

The most important assumptions of binary logistic regression are:
* The dependent variable is a binary categorical variable.
* The independent variables can be continuous or categorical.
* The log odds of the dependent variable are a linear combination of the independent variables.
* The log odds follow a logistic distribution.
* The independent variables are independent of each other.
* There is little or no multicollinearity among the independent variables.
* The sample size is large enough to ensure stable estimates of the parameters.

#### Read data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import make_blobs

sns.set_style('darkgrid')
```

```python
X, y = make_blobs(n_samples = 500, centers = 2, cluster_std = 1, n_features = 2, random_state=30)
df = pd.concat([pd.DataFrame(X).rename(columns={0:'feature1', 1: 'feature2'}), 
                pd.Series(y).rename('label')], axis=1)
df.head()
```

```python
plt.figure(figsize=(7, 5))
plt.title('Scatterplot of generated data')
ax = sns.scatterplot(data=df, x='feature1', y='feature2', hue='label')
ax.legend(labels={'Group 1', 'Group 2'})
plt.show()
```

#### Sigmoid function

A binary logistic regression takes in an input (in this case feature1 and feature2) and returns a probability, a value between 0 and 1. <br> 
How does it do that? Using the sigmoid function shown below:

$$\phi(z) = \frac{1}{1+e^{-z}} \text{  where } z = \theta^T X$$

This function maps real numbers to a values between 0 and 1.

```python
def sigmoid(X, theta): return 1 / (1 + np.exp(-np.dot(X, theta)))
```

```python
x = np.arange(-10, 10, 0.1)
y = 1 / (1 + np.exp(-x))

plt.plot(x, y, color='blue')
plt.axhline(y=0, color='grey', linestyle='-')
plt.axvline(x=0, color='grey', linestyle='-')
plt.show()
```

From here, there are two common ways to approach the optimization of the Logistic Regression. One is through loss minimizing with the use of gradient descent and the other is with the use of Maximum Likelihood Estimation.

#### Loss function

The loss function used in binary logistic regression is the cross-entropy loss function, defined as:

$$J(\theta) = -\frac{1}{N}(y \times \log(h) + (1 - y) \times \log(1 - h)) \text{ where } h = sigmoid(\theta^T X)$$

To clarify further, y is the true label (0 or 1) and p is the predicted probability of the positive class. The cross-entropy loss is a measure of how well the model's predictions match the true labels.

```python
def loss(h, y): return -(y * np.log(h) + (1-y) * np.log(1-h)).mean()
```

#### Gradient Descent

The goal is to minimize the loss by increasing or decreasing the weights, which will be decided by a function called Gradient descent. <br>
Gradient descent is an iterative optimization algorithm to find the local minimum of a function.<br>

<img src="Images/gradient_descent.png" width="400"></img>

The local minimum of a function is found by taking steps towards the negative gradient of the function at the current point, as the gradient points at the direction of steepest ascent.<br>

Therefore, we will update the weights of the features following the formula below:

$$
\begin{split}
    \theta &:= \theta - \alpha \times \frac{\partial(J(\theta))}{\partial(\theta)} \\ 
    &:= \theta - \alpha \times \frac{1}{N} X^T(h-y) \\ 
\end{split}
$$

where $\alpha$ is the learning rate and $\frac{\partial(J(\theta))}{\partial(\theta)}$ is the gradient.

```python
def calc_gradient(X, h, y): return np.dot(X.T, h - y) / X.shape[0]

def update_weights(theta, learning_rate, gradient): return theta - learning_rate * gradient
```

#### Minimize the loss function using Gradient Descent

```python
# Add intercept
intercept = np.ones((X.shape[0], 1)) 
X = np.concatenate((intercept, X), axis=1)
```

```python
theta = np.zeros(X.shape[1]) # Initialize weights as 0
loss_list = []

for i in range(10000):
    #Apply sigmoid function to logistic regression
    h = sigmoid(X, theta)
    
    # Calculate loss
    loss_list.append(loss(h, y))
    
    # Calculate the gradient and update the weights
    gradient = calc_gradient(X, h, y)
    theta = update_weights(theta=theta, learning_rate=0.1, gradient=gradient)
```

```python
plt.figure(figsize=(12, 5))
plt.title('Cost function J')

plt.plot(loss_list)
plt.ylim([0, 0.8])
plt.ylabel('Cost')
plt.xlabel('No. of iterations')
plt.show()
```

```python
print('The optimal theta values are: {}'.format(theta))
```

#### Plot decision boundary

```python
plt.figure(figsize=(7,5))

# Retrieve the model parameters.
b, w1, w2 = theta[0], theta[1], theta[2]

# Calculate the intercept and gradient of the decision boundary.
c = -b/w2
m = -w1/w2

# Plot the data and the classification with the decision boundary.
xmin, xmax = -1, 7
ymin, ymax = -10, 3
xd = np.array([xmin, xmax])
yd = m*xd + c
plt.plot(xd, yd, 'k', lw=1, ls='--')
plt.fill_between(xd, yd, ymin, color='tab:orange', alpha=0.2)
plt.fill_between(xd, yd, ymax, color='tab:blue', alpha=0.2)

plt.title('Decision boundary')
sns.scatterplot(x=df[df['label'] == 0]['feature1'], y=df[df['label'] == 0]['feature2'])
sns.scatterplot(x=df[df['label'] == 1]['feature1'], y=df[df['label'] == 1]['feature2'])
plt.xlim(xmin, xmax)
plt.ylim(ymin, ymax)
plt.ylabel('feature2')
plt.xlabel('feature1')

plt.show()
```
